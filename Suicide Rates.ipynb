# use suicides/100k pop to analysis

```python
country = df[['country','suicides/100k pop']].groupby('country',as_index=False).sum().sort_values(by='suicides/100k pop',ascending=False).head(10)
sns.barplot(x='country',y='suicides/100k pop',data=country)
plt.xticks(rotation=90)
plt.title('Top 10 countries in suicides/100k pop')
```

# Top 3 Lithuania,Hungary,Russian Federation 

```python
top3_100k_pop=df.loc[(df.country == 'Lithuania')|(df.country == 'Hungary')|(df.country == 'Russian Federation')]
```

```python
top3_100k_pop.head()
```

```python
corr = top3_100k_pop.corr()
plt.figure(figsize = (20,7))
sns.heatmap(corr, annot = True)
plt.title('heatmap')
```

# population(Top3 suicides_no & dataset)

```python
fig, axe1 = plt.subplots()
axe2 = axe1.twinx()
c1 = sns.scatterplot(ax = axe1, data = df, x = 'population', y = 'suicides_no', color = 'red')
c2 = sns.scatterplot(ax = axe2, data = top3_100k_pop, x = 'population', y = 'suicides_no', color = 'blue')
axe1.set_ylabel('dataset suicides_no',color='red')
axe2.set_ylabel('top3 suicides_no',color='blue')
plt.title('population(Top3 suicides_no & dataset)')
plt.show()
```

# gdp_per_capita (Top3 suicides_no & dataset)

```python
fig, axe1 = plt.subplots()
axe2 = axe1.twinx()
c1 = sns.scatterplot(ax = axe1, data = df, x = 'gdp_per_capita ($)', y = 'suicides_no', color = 'red')
c2 = sns.scatterplot(ax = axe2, data = top3_100k_pop, x = 'gdp_per_capita ($)', y = 'suicides_no', color = 'blue')
axe1.set_ylabel('dataset suicides_no',color='red')
axe2.set_ylabel('top3 suicides_no',color='blue')
plt.title('gdp_per_capita (Top3 suicides_no & dataset)')
plt.show()
```

# HDI for year(Top3 suicides_no & dataset)

```python
fig, axe1 = plt.subplots()
axe2 = axe1.twinx()
c1 = sns.scatterplot(ax = axe1, data = df, x = 'HDI for year', y = 'suicides_no', color = 'red')
c2 = sns.scatterplot(ax = axe2, data = top3_100k_pop, x = 'HDI for year', y = 'suicides_no', color = 'blue')
axe1.set_ylabel('dataset suicides_no',color='red')
axe2.set_ylabel('top3 suicides_no',color='blue')
plt.title('HDI for year (Top3 suicides_no & dataset)')
plt.show()
```
